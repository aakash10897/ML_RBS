### Step 6: Create a pie chart to show ratio among different modes of benefits

```python
df.columns
```

```python
df.columns=['Year','Elementary','Adult','Higher','Computer','total']
```

```python
elementary=df['Elementary'].sum(axis=0)
adult=df['Adult'].sum(axis=0)
higher=df['Higher'].sum(axis=0)
computer=df['Computer'].sum(axis=0)
```

```python
total_benefits=pd.Series({'elementary':elementary,'adult':adult,'higher':higher,'computer':computer})
```

```python
total_benefits
```

```python
plt.figure(figsize=(15, 10))
plt.pie(total_benefits, labels=total_benefits.index, 
        autopct="%.2f")
```

### Step 7: For the 5 highest benefitted states, create 
1. Bar chart. 
2. Stacked bar chart 
3. 100% stacked chart

```python
df.drop('Year',axis=1,inplace=True)
```

```python
df.head(5)
```

```python
highest_states=df.sort_values('total',ascending=False).head(5)
```

```python
highest_states
```

```python
highest_states.drop('total',axis=1,inplace=True)
```

```python
highest_states.plot(kind='bar')
```

```python
highest_states.plot(kind='bar',stacked=True)
```

```python
highest_states['total']=highest_states.sum(axis=1)
```

```python
highest_states
```

```python
highest_states = highest_states.apply(lambda a:round(100*a/highest_states['total']))
```

```python
highest_states
```

```python
highest_states.plot(kind='bar',stacked=True)
```
