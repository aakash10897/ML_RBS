```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sea
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
train_data=pd.read_csv('data/train.csv')
test_data=pd.read_csv('data/test.csv')
```

```python
sea.set()
plt.figure(figsize=(15,15))
sea.heatmap(train_data.corr(),annot=True,linewidth=1,linecolor='w')
```

```python
training_data=train_data[['PassengerId','Pclass','Fare']]
output_data=train_data['Survived']
```

```python
x,x_test,y,y_test=train_test_split(training_data,output_data,test_size=0.3,random_state=100)
model =LogisticRegression()
```

```python
model.fit(x,y)
```

```python
prediction=model.predict(x_test)
prediction
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
confusion_matrix(y_test,prediction)
```

```python
accuracy_score(y_test,prediction)
```

# Making new prediction

```python
gender_dict ={
    'male' : 0,
    'female' : 1
}
```

```python
train_data['Sex'].unique()
```

```python
train_data['Sex'] = train_data['Sex'].map(gender_dict)
```

```python
plt.figure(figsize=(15,15))
sea.heatmap(train_data.corr(),annot=True,linewidth=1,linecolor='w')
```

```python
training_data=train_data[['PassengerId','Pclass','Fare','Sex']]
```

```python
output_data=train_data['Survived']
x,x_test,y,y_test=train_test_split(training_data,output_data,test_size=0.3,random_state=42)
```

```python
model=LogisticRegression()
model.fit(x,y)
```

```python
prediction=model.predict(x_test)
```

```python
confusion_matrix(y_test,prediction)
```

```python
accuracy_score(y_test,prediction)
```

# Importing test data

```python
test_data=pd.read_csv('data/test.csv')
```

```python
test_input=test_data[['PassengerId','Pclass','Fare','Sex']]
test_input['Sex']=test_input['Sex'].map(gender_dict)
test_input.head(2)
test_input['Fare']= test_input['Fare'].fillna(test_data['Fare'].mean())
```

```python
prediction=model.predict(test_input)
```

```python
prediction_df=pd.DataFrame({'PassengerId':test_input['PassengerId'],'Survived':prediction})
```

```python
prediction_df.set_index('PassengerId',inplace=True)
```

```python
prediction_df.to_csv('Prediction.csv')
```
